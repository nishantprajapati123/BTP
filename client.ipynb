```python
import pandas as pd
```

```python
df = pd.DataFrame()
df = pd.read_csv("train_proper_csv_attack_types.csv",header = None)
```

```python
df.head()
```

```python
import weka.core.serialization as serialization
from weka.classifiers import Classifier
objects = serialization.read_all("j48_80split.model")
classifier = Classifier(jobject=objects[0])
print(classifier)
```

```python
df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[0])
```

```python
df = df.drop('unknown2',axis=1)
```

```python
df_dummies = pd.get_dummies(df,columns = ["protocol_type","service","flag"])
```

```python
len(list(df_dummies))
```

```python
df_dummies
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df_tr = df_dummies.drop('unknown1',axis = 1)
from sklearn.feature_selection import SelectFpr
from sklearn.feature_selection import chi2

x_new = SelectFpr(chi2, alpha=0.05).fit_transform(df_tr, df_dummies['unknown1'])

clf = RandomForestClassifier(n_estimators=10)
clf.fit(x_new,df_dummies['unknown1'])
```

```python
df1 = pd.DataFrame()
df1 = pd.read_csv("test_proper_csv_attack_types.csv",header = None)
```

```python
df1.head()
```

```python
df1 = df1.rename(columns=df1.iloc[0])
df1 = df1.drop(df1.index[0])
df1 = df1.drop('unknown2',axis=1)
df1_dummies = pd.get_dummies(df1,columns = ["protocol_type","service","flag"])
```

```python
df1_dummies['service_aol']=0
df1_dummies['service_harvest']=0
df1_dummies['service_http_2784']=0
df1_dummies['service_http_8001']=0
df1_dummies['service_red_i']=0
df1_dummies['service_tftp_u']=0
df1_dummies['service_urh_i']=0
```

```python
for item in list(df_dummies):
    if(item not in list(df1_dummies)):
        print (item)
```

```python
df1_ts = df1_dummies.drop('unknown1',axis = 1)
```

```python
df1_out = clf.predict(df1_ts)
```

```python
df1_out
```

```python
from sklearn.metrics import classification_report
print(classification_report(df1_dummies['unknown1'],df1_out))
```
